## Introduction
> Torch Tensors vs Numpy Array
- torch tensor are similar to numpy ndarrays and can share the same underlying memory except torch tensors are optimized for automatic differentation

## Initializing a tensor

```python
import torch
import numpy as np
```

```python
# Direct from data
data = [[1,2],[3,4]]
x_data = torch.tensor(data, dtype=torch.int32)
print(x_data)
```

```python
# From numpy array
np_data = np.array(data) # create np ndarray
x_np = torch.from_numpy(np_data) # create torch tensor from np ndarray
print(x_np)
```

```python
# From tensors
x_ones = torch.ones_like(x_data) # retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")
x_rand = torch.rand_like(x_data, dtype=torch.float32) # overrides the datatype of x
print(f"Random Tensor: \n {x_rand} \n")
```

```python
# Random or constant initialization
shape = (5,10)
x_rand = torch.rand(shape, dtype=torch.float32) # uniform random values between 0 and 1
x_ones = torch.ones(shape)
x_zeros = torch.zeros(shape)
x_randint = torch.randint(low=0, high=10, size=shape) # random integers from low to high
```

## Tensor Attributes
> Among the more important ones are: `shape`, `dtype`, `device`

```python
print(x_data.shape)
print(x_data.dtype)
print(x_data.device) 
```

## Tensor Operations

```python
# Set device agnostic code,
# we move our tensors to accelerator if available, otherwise we use CPU
device = torch.accelerator.current_accelerator() if torch.accelerator.is_available() else "cpu"
x = torch.rand(3,3).to(device)
x
```

```python
# Splicing
print(x[1:, 1:])
```

```python
# Joining tensors with torch.cat()
```

```python
x2 = torch.rand(4,3).to(device)
x3 = torch.cat([x, x2], dim=0) # concatenate along rows 
print(x3, x3.shape)

x4 = torch.rand(3,4).to(device)
x5 = torch.cat([x, x4], dim=1) # concatenate along columns
print(x5, x5.shape)
```
