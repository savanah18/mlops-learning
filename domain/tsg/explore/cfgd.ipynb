# Background

## Theory
### Variance Preserving Markov Process (Sohl-Dickstein et. al., 2015)
Let 
- $x \sim p(x)$
- $z = \{z_\lambda | \lambda \in [\lambda_{min}, \lambda_{max}]  \}$

### Forward
$q(z_\lambda | x) = \mathcal{N}(\alpha_\lambda x , \sigma_\lambda^2 I)$ where
- $\alpha_\lambda^2 = \frac{1}{1+e^{-\lambda}}$ - sigmoidal
- $\sigma_\lambda^2 = 1 - \alpha_\lambda^2$

### Reverse
$q(z_\lambda | z_{\lambda^\prime}) = \mathcal{N}(\frac{\alpha_\lambda}{\alpha_{\lambda^\prime}}z_{\lambda^\prime}, \sigma_{\lambda|\lambda^\prime}^2I)$ where
- $\lambda \lt \lambda^\prime$
- $\sigma_{\lambda|\lambda^\prime}^2 = (1-e^{\lambda - \lambda^\prime})\sigma_\lambda^2$

> Note: $\lambda=\log\frac{\alpha_\lambda^2}{\sigma_\lambda^2}$  can be interpreted as signal to noise ratio (SNR)
